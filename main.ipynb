### Simulation Data Generation and Data Rationality Consideration

```python
with open(file=r'C:\vscode_c\vscode_cc\code\dataset\milk_tea\original.csv', mode='r', encoding='utf-8') as f:
    original = f.readlines()
    original = [x.split('\t') for x in original]
original.pop(0)
```

```python
# original
# original reprocessed data
```

```python
user_id_list = [x[0] for x in original]
# user_id_list: list with all user_id
user_id_set = list(set(user_id_list))
# user_id_set: list with user_id not duplicated
user_id_len = len(user_id_set)
user_id_dict = dict()
for i in range(user_id_len):
    user_id_dict[user_id_set[i]] = i + 1
# user_id_dict: user_id map to uid
# len(user_id_set) = 1210271
```

```python
item_id_list = [x[1] for x in original]
# item_id_list: list with all item_id
item_id_set = list(set(item_id_list))
# item_id_set: list with item_id not duplicated
item_id_len = len(item_id_set)
item_id_dict = dict()
for i in range(item_id_len):
    item_id_dict[item_id_set[i]] = i + 1
# item_id_dict: item_id map to iid
# len(item_id_set) = 249274
```

```python
milk_tea = [[user_id_dict[user_id], item_id_dict[item_id], rating, timestamp] for user_id, item_id, rating, timestamp in original]
with open(file=r'C:\vscode_c\vscode_cc\code\dataset\milk_tea\milk_tea.inter', mode='w', newline='',encoding='utf-8') as csv_f:
    csv_f.write('user_id:token' + '\t' + 'item_id:token' + '\t' + 'rating:float' + '\t' + 'timestamp:float' + '\n')
    for line in milk_tea:
        csv_f.write(str(line[0]) + '\t' + str(line[1]) + '\t' + str(line[2]) + '\t' + str(line[3]) + '\n')
# generate milk_tea.inter file
```

```python
from collections import Counter, defaultdict

iid_list = [x[1] for x in milk_tea]
iid_count = dict(Counter(iid_list))
# iid map to frequency

categories = 10
boundaries = list(range(1, item_id_len + 1, round(item_id_len / categories) + 1))
boundaries.append(item_id_len + 1)
boundaries_dict = dict()
for i in range(categories):
    for j in range(boundaries[i], boundaries[i + 1]):
        boundaries_dict[j] = i + 1
# 249274: divided into 10 categories
# 10 intervals: [a, b)
# iid map to category

category_count = defaultdict(lambda: 0)
for iid, frequency in iid_count.items():
    category_count[boundaries_dict[iid]] += frequency
# amount of each category
# even distribution, in line with reality

category_count
```

```python
import pyecharts.options as opts
from pyecharts.charts import Line

category_keys = list(category_count.keys())
x = ['category_' + str(i) for i in category_keys]
y = [category_count[i] for i in category_keys]
line=(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name="item number",y_axis=y, is_smooth=True)
    .set_global_opts(title_opts=opts.TitleOpts(title="Iid Number of Categories Involved"))
)
# line.render_notebook()
line.render("iid_number.html")
```

```python
uid_category = defaultdict(lambda: defaultdict(lambda: 0))
boundaries_dict[iid] += 1
for uid, iid, rating, timestamp in milk_tea:
    uid_category[uid][boundaries_dict[iid]] += 1
# uid map to a dict, which contains amount of each category involved

uid_category
```

```python
uid_category_count = {uid: len(boundary_dict) for uid, boundary_dict in uid_category.items()}
# uid map to number of categories involved

uid_category_count
```

```python
x = list(uid_category_count.keys())
y = list(uid_category_count.values())
x = x[:50]
y = y[:50]
x = [str(i) for i in x]
y = [str(i) for i in y]
line=(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name="number of categories involved",y_axis=y, is_smooth=True)
    .set_global_opts(title_opts=opts.TitleOpts(title="Category Number of Uid (Part)"))
)
# line.render_notebook()
line.render("uid_category_count.html")
```

```python
category_uid_count = defaultdict(lambda: 0)
for category_count in uid_category_count.values():
    category_uid_count[category_count] += 1
# number of categories involved map to uid number
# decreases exponentially, in line with reality
```

```python
x = ['category_' + str(i + 1) for i in list(range(categories))]
y = list(category_uid_count.values())
y.sort(reverse=True)
line=(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name="user number",y_axis=y, is_smooth=True)
    .set_global_opts(title_opts=opts.TitleOpts(title="Uid Number of Categories Involved"))
)
# line.render_notebook()
line.render("uid_number.html")
```
